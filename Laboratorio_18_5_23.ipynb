```python
import folium

from folium.plugins import minimap
Heredia = folium.Map(location=(9.958291185423821, -84.13800976410099))
Heredia
```

```python
from folium.plugins import MiniMap
from branca.element import Figure

popuptext = 'Heredia'
my_map = folium.Map(location=[9.970692652682622, -84.129321590024], zoom_start=16, tiles='stamenterrain')
folium.Marker(location=[9.970692652682622, -84.129321590024], popup=popuptext, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)

folium.Circle(location=[9.970692652682622, -84.129321590024], color="purple", weight=6, fill_color="red", fill_opacity=0.5, tooltop="Heredia").add_to(my_map)

minimap= MiniMap(tile_layer='stamenterrain', position = 'bottomright')

my_map.add_child(minimap)

my_map.save('C:folium.html')

my_map
```

```python
fig=Figure(width=1000, height=500)
fig.add_child(my_map)
fig
```
